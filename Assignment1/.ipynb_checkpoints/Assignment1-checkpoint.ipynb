## 1) Problem: Adjacency matrix of Erdös-Rényi graph

```julia
using RDatasets
srand(1655244)
function ER(n,p)
    # input: n = size of the square matrix
    # input: p = probability that each entry of the output matrix is 1
    # output: A = adjacency matrix of the Erdos-Renyi matrix
    if( n <= 0 )
        println("n must be a positive value")
    elseif ( p > 1 || p < 0)
        println("p must be between 0 and 1")
    else
        A = rand(n*n)
        A[A.<=p] = 1 
        A = floor(A) # floor(1) = 1, all the others will be 0
        A = reshape(A, n, n) # build the matrix
        return(A)
    end 
end
```

```julia
# Examples
A = ER(5, .3);
B = ER(5, .7);
```

## 2) Problem: Union of two adjacency matrices

```julia
function Union(A, B)
    # input: A, B adjacency matrix of size n x n. 
    # output: C adjacency matrix with the union of edges of A and B
    if(size(A) != size(B))
        println("matrix must be of the same size")
    else
        C = A+B # the sum may contain cells equal to 2
        C[C.>0] = 1 # then, just transform all positive values to 1
        return C
    end
end
```

```julia
AuB = Union(A,B);
```

## 3) Problem: Adjacency matrix of cycle graph

```julia
function Cycle(n)
    # input n: size of output matrix (n x n)
    # output A: adjacency matrix of the cycle graph
    A = zeros(n, n)
    for i = 1:(n-1)  
        A[i, i+1] = 1
    end
    A[n,1] = 1
    return A
end
```

```julia
Cycle(5)
```

## 4) Problem: A from B function

```julia
# For this function, only one row per edge will be considered, I considered this way 
# because in the restriction X_j = X_i  is equal to  X_i = X_j. Then, less rows to 
# sample in the algorithm!
function AfromB(B)    
    # input B: adjacency matrix of size nxn
    # output A: matrix with 1 row corresponding to 1 edge in the adjacency matrix B
    n = size(B)[1]
    m = convert(Integer, sum(B))
    A = zeros(m, n)
    row = 0
    for i = 1:n
        for j = 1:n
            if B[i, j] == 1
                row += 1
                A[row, j] = 1
                A[row, i] = -1
            end
        end
    end
    return A
end
```

```julia
AfromB(B)
```

## 5) Problem: Rate ($\rho$) of the randomized Kaczmarz method

```julia
function rate(A)
    # input A: matrix generated from the method AfromB
    # output B: rate (\rho) 
    eigvalues = sort(eig(A'A)[1])
    rho = 1 - eigvalues[eigvalues .> .00000001][1]/(vecnorm(A, 2)^2) # precision of the numerical "0" CAN be modified 
    # according to the specified precision.
    return rho
end
```

```julia
rate(A)
```

## 7) Problem: Numerical experiment for different probabilities for ER function

```julia
n = 100
p = linspace(0.01, 1, 100)
values = zeros(1,100)
rho = zeros(1,100)

for i=1:100
    rho[i] = rate(AfromB(Union(Cycle(n), ER(n, p[i]))))
    values[i] = rho[i]/(1-rho[i])
end
```

```julia
#using Gadfly for plots 
using Gadfly
values_df = convert(DataFrame, values')
values_df2 = convert(DataFrame, rho')
values_df = DataFrame(x = (1:100)/100, y = values_df[:,1], rho = values_df2[:,1])

plot(values_df, x=:x, y=:y, 
    Geom.point, Geom.smooth(method=:loess), 
    Guide.title("rho/(1-rho)"), Guide.xlabel("Probability of the ER"), Guide.ylabel("Value of rho/(1-rho)"))
```

This graph shows that, if the probability of the ER function is close to 1, then the value of $\frac{rho}{1-rho}$ converges to the 19899. In other cases it seems that the value decreases with respect to the probability. Due to the randomness of the matrix, it is important to analize how is the distribution under different simulations. This is analyzed below.

### Analyzing extreme cases:

First the two extreme cases are analized. This two examples are not random because the probability takes or 0 or 1. After this, simulations with distinct probabilities (.05, .1, .2, .5, .7) will be made.

#### 1) Case probability = 0
The first extreme case that will be analized is when the probability for the ER function is 0. Then, all the entries of the ER adjacency matrix will be 0: (in this report the name of odd rate will be assigned to $\frac{\rho}{1-\rho}$ in inspiration of the odds ratio in probability)

```julia
# Case 1: p = 0
A = AfromB(Cycle(n))
eigA = eig(A'A)[1]
println("\t The eigenvalues of A'A are: " )
println(eigA)
rho_ex = 1-0.00394654/200
rho_odds = rho_ex/(1-rho_ex)
@printf "\t The sum of the eigenvalues is: %f" sum(eigA) 
@printf "\n\t The rate for this matrix is: %f" rho_ex
@printf "\n\t The odds rate for this matrix is: %f" rho_odds
```

Only one eigenvalue of this matrix is 0, all the others are positive. The $\lambda_{min}^+$ is **0.00394654**. Also, the sum of all eigenvalues is **200**. Then the rate is $1 - \frac{0.00394654}{200} = 0.999980$

#### 2) Case probability = 1
The second extreme case that will be analized is when the probability for the ER function is 1. Then, all the entries of the ER adjacency matrix will be 1:

```julia
# Case 2: p = 1
A = AfromB(Union(Cycle(n), ER(n, 1)))
eigA = eig(A'A)[1]
println("\t The eigenvalues of A'A are: " )
println(eigA)
rho_ex = 1-1/19900
rho_odds = rho_ex/(1-rho_ex)
@printf "\t The sum of the eigenvalues is: %f" sum(eigA) 
@printf "\n\t The rate for this matrix is: %f" rho_ex
@printf "\n\t The odds rate for this matrix is: %f" rho_odds
```

Any eigenvalue of this matrix is 0. The $\lambda_{min}^+$ is **1**. Also, the sum of all eigenvalues is **19900**. Then the rate is $1 - \frac{1}{19900} = .99995$

#### Comparing distributions of rates given different probabilites.

```julia
case1_rate = rate(AfromB(Union(Cycle(n), ER(n, 0))))
case2_rate = rate(AfromB(Union(Cycle(n), ER(n, 1))))
@printf "\t odds rate of the first extreme case: %f" case1_rate/(1-case1_rate)
@printf "\n\t odds rate of the second extreme case: %f" case2_rate/(1-case2_rate)

vector_probs = [.05, .1, .2, .5, .7]
case3_rate_odds = zeros(500,6)
for i = 1:500
    for j = 1:5
        x = rate(AfromB(Union(Cycle(n), ER(n, vector_probs[j]))))
        case3_rate_odds[i, j] = x/(1-x)
        case3_rate_odds[i, 6] = i
    end
end
```

```julia
case3_rate_df = convert(DataFrame, case3_rate_odds)
rename!(case3_rate_df, 
    [:x1 => :p05, 
    :x2 => :p10,
    :x3 => :p20,
    :x4 => :p50,
    :x5 => :p70])
case3_rate_df = melt(case3_rate_df, :x6)
plot(case3_rate_df, x=:value, color=:variable, xintercept=[19899.000000, 50676.261778],
    Geom.density, Guide.title("Simulations for distinct probabilities"), 
    Guide.xlabel("Value of 'odd rate'"), Guide.ylabel("Density"), Geom.vline)
```

The vertical blue lines are the extreme cases and the color distributions are the distributions under distinct probabilities. From this graph is easy to see how when p increases, then the distribution of the $\frac{\rho}{1-\rho}$ is coverging to the value when p = 1. Although, is easy to see from the distributions, that some cases behave better than the two extreme cases. This means that an intelligent allocation of values in the original matrix can have better values of $\rho$
