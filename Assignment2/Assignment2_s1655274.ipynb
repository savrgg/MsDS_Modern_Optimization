### Transform the original to problem to an easier formulation
In the original problem:
$$f(x) = \left[\; \sum_{i = 1}^{d} \phi_j^*(b_j^T x) \;\right] + \gamma x^T x$$

f(x) can be rewritten in a easier way, but some considerations about the function $\phi_j^*(s)$ should be done. This is a piecewise function, so it is necessary to consider all the cases. First, we will consider the case when s<-1: 

$$f(x) = \left[\; \sum_{i = 1}^{d} \frac{(b_j^T x + 1)^2}{2\alpha_j} \;\right] + \gamma x^T x$$
$$f(x) = \left[\; \frac{1}{2}\sum_{i = 1}^{d} \left(\;\frac{b_j^T}{\sqrt{\alpha_j}} x + \frac{1}{\sqrt{\alpha_j}}\;\right)^2 \;\right] + \gamma x^t x$$

Now, let's call $A_j = \frac{b_j^T}{\sqrt{\alpha_j}}$ and $c_j = \frac{1}{\sqrt{\alpha_j}}$. (for the other cases, c will be defined different. For $s>1$, $c_j = -\frac{1}{\sqrt{\alpha_j}}$ and for $s \in (-1,1)$, $c_j = -A_j^Tx$)
$$f(x) = \left[\; \frac{1}{2}\sum_{i = 1}^{d} \left(\;A_j x + c_j\;\right)^2 \;\right] + \gamma x^T x$$

This last expression is equivalent to the norm $||Ax||^2_2 = \sum_j(A_jx)^2$ with $A_j$ the rows of A:

$$f(x) = \frac{1}{2}||Ax + c||^2_2  + \gamma ||x^Tx||$$

$$f(x) = \frac{1}{2}(Ax + c)^T (Ax + c)  + \gamma x^T x $$
$$f(x) = \frac{1}{2}\left(\;\ x^TA^TAx +  2 x^TA^Tc + c^Tc\;\right) + \gamma x^T x $$

Taking the derivative of the last expression:
$$\nabla f(x) = x^TA^TA + c^TA + 2\gamma x^T$$

Now, using the Definition 8 from the lecture 7, we have that $f(x+h) \leq f(x) + \nabla f(x)^T h + h^T M h$. Let's divide this expression in LHS and RHS. The LHS is equivalent to:

$$f(x+h) = \frac{1}{2} (x+h)^TA^TA(x+h) +  (x+h)^TA^Tc + \frac{1}{2}  c^Tc + \gamma (x+h)^T(x+h) $$
$$f(x+h) = \left(\;\ \frac{1}{2} x^TA^TAx +  x^TA^Tc + \frac{1}{2}  c^Tc + \gamma x^Tx \;\right) + \left(\;\ x^TA^TAh + c^TAh + 2\gamma x^Th \;\right)+ h^T (A^TA + \gamma I)h$$ 

And the RHS:
$$f(x) + \nabla f(x)^T h + h^T M h =\left(\;\ \frac{1}{2} x^TA^TAx +  x^TA^Tc + \frac{1}{2}  c^Tc + \gamma x^Tx \;\right) + \left(\;\ x^TA^TAh + c^TAh + 2\gamma x^Th \;\right)+ h^T M h$$

This inequality is an equality when $M = A^T A + \gamma I $. This result makes sense with the theorem 9 from the lecture 7. The dimmensionalities of the matrices and vectors for this assignment are the following:

$A \in ℝ^{d\times n} \qquad M \in ℝ^{n\times n} \qquad x \in ℝ^{n\times 1} \qquad c \in ℝ^{d\times 1} \qquad h \in ℝ^{n\times 1} \qquad v \in ℝ^{n\times 1}$

Now, all the theorems in the class works with a factorization of M like $M = A^T A$. So, the first idea tried was to use Cholesky factorization of M. The problem here, is that each factor A for the Cholesky factorization is a $n\times n$ matrix, so if n>>d, our computations are much much expensive (Remember that we make iterations over the matrix A). In consequence, a second idea is given below.

### Include penalization matrix in the original matrix A

In the first section the problem was defined as:

$$f(x) = \frac{1}{2}||Ax + c||^2_2  + \gamma ||x^Tx||$$

With each row of A equal to $A_j = \frac{b_j^T}{\sqrt{\alpha_j}}$ and each element of c equal to $c_j = \frac{1}{\sqrt{\alpha_j}}$. In machine learning this optimization problem can be seen as an augmented problem, with $D = {A \choose \sqrt{\gamma}I}$ (Add rows to the matrix A corresponding to the identity $n\times n$) and $e = {c \choose 0}$ (with c the original vector c and 0 all the others entries):

$$f(x) = \frac{1}{2} ||Dx + e||^2_2$$

$$f(x) = \frac{1}{2} (Dx+e)^T (Dx+e)$$

$$f(x) = \frac{1}{2} x^TD^TDx + x^TD^Te + \frac{1}{2} e^T e$$

With these formulations and using multiplication by blocks, $D^TD$ is equivalent to $(A \sqrt{\gamma}I) {A \choose \sqrt{\gamma}I} = A^T A + \gamma I $, $ x^T D^T e = x^T(A \sqrt{\gamma}I) {c \choose 0} = x^TA^Tc $ and e^T e = c^T c. So:

$$f(x) = \frac{1}{2} x^T (A^T A + \gamma I) x + x^TA^Tc + \frac{1}{2} c^T c$$
$$f(x) = \frac{1}{2} x^T (A^T A)x + x^T (\gamma I) x + x^TA^Tc + \frac{1}{2} c^T c$$
$$f(x) = \frac{1}{2} (Ax +c)^T (Ax+c) + \gamma x^Tx $$

Then, both problems are equivalent. Now, we should inspect the behaviour of the other parameters. To calculate the $v_i$'s we need to compute $\omega_j$, but with this new formulation, $D \in R^{(n+d) \times n}$. Consequently we need $(n+d)$ $\omega_j$!!! 

Fortunately, all of the $\omega_j$ from $j = d+1$ to $n+d$ are 1 (because of the diagonal matrix $\sqrt{\gamma} I$). So, in the formula (36) of lecture seven, the factor $\frac{(\omega_j -1)( \tau-1)}{max(n-1, 1)}$ becomes 0 for all the $j$ in this range. Now, the sum in formula (36) from $j = d+1$ to $n+d$ is simplified to sum just $A_{ji}^2$. Because of the diagonal matrix in this block of the matrix, just one summand is added to the sum, and it is the one in the diagonal: $\gamma$. As a consequence, it is possible just to iterate over the original matrix A and just add gamma \times x^x at the end.



```julia
srand(123456)

# Define parameters for the function ComputeESO ----------------------
global n = 1000              # number of columns of A
global d = 10                # number of rows of A
global B = sprandn(d,n,0.9)  # Create random normal matrix with distinct level of sparsity
global alpha = zeros(d,1)    # useful to calculate A, c_1 and c_3
global gamma = 1/d           # penalization term
global tau = 2               # parameter of the T-nice

# global A, c_1 and c_3 according to the above definitions------------
global c_1 = zeros(d,1)      # for new_nomenclature (for the case s <-1)
global c_3 = zeros(d,1)      # for new_nomenclature (for the case s > 1)
                             # for new_nomenclature (the case s \in (-1,1) just put 0)
global A = zeros(d,n);       # for new_nomenclature


# To implement T-nice sampling, we need to sample randomly from the
# distribution of the combinations. I use the samplinK function of 
# the professor to do this.
function samplingK(k, n) 
    reservoir = collect(1:k)
    for i=(k+1):n
        pos = rand(1:i)
        if pos <= k
            reservoir[pos] = i
        end
    end
    return reservoir
end
```

## 1) Problem 1: ComputeEso

```julia
function ComputeESO(B, alpha, gamma, tau)
    # First of all, compute A, c_1, c_3 for the new nomenclature------
    for i = 1:d
        alpha[i] = n/i                   # Given parameter 
        c_1[i] = 1/(sqrt(alpha[i]))      # calculate c_j for the case 1 of \phi^*
        c_3[i] = -1/(sqrt(alpha[i]))     # calculate c_j for the case 3 of \phi^*
        A[i,:] = copy(B[i,:])            # careful when assigning a vector to another vector!
        A[i,:] = A[i,:]/sqrt(alpha[i])   # With the new definitions, scale the input matrix
    end
    
    # Now, let's create the omegas and v ------------------------------
    omega = zeros(d,1)                   # Empty array of omegas
    v = zeros(n,1)                       # Empty array for v
    
    # Omega computation -----------------------------------------------
    for j = 1:d                          # The matrix A is d \times n
        for i = 1:n
            if A[j,i]!=0
                # According to the formula of lecture 7
                omega[j] += 1            # per row, count number of columns!=0
            end
        end
    end
    # v computation ---------------------------------------------------
    for i = 1:n                          # The matrix is n \times n
        for j = 1:d
            # According to the formula of lecture 7
            v[i] = v[i]+(1+(tau-1)*(omega[j]-1)/(n-1))*A[j,i]*A[j,i]
        end
        v[i] = v[i] + 2*gamma            # Add the 2 times lambda of the above derivation
    end
    return v
end
```

For A dense, $v_i$ is equivalent to $\tau M_{ii}$

For A sparse, $v_i$ is equivalent to $\omega_{max} M_{ii}$

```julia
# Compute v parameters for tau-nice ----------------------------------
println(" The norm of v using the formulation for tau-nice is:")
for i = 1:3
    tau = i
    v = ComputeESO(B, alpha, gamma, tau)
    xx = norm(v)
    println("For tau = $tau the norm of v is $xx")
end

# Compute v parameters for dense matrix ----------------------------------
println("\n The norm of v using the formulation for dense matrix is:")
for i = 1:3
    tau = i
    M = A'*A+ 2*gamma*eye(n)
    v_dense = zeros(n,1)
    for i = 1:n
        v_dense[i] = tau*diag(M)[i]      # formula (33)
    end
    xx = norm(v_dense)
    println("For tau = $tau the norm of v is $xx")
end

# Compute v parameters for sparse matrix ----------------------------------
println("\n The norm of v using the formulation for sparse matrix is:")
omega = zeros(d,1)
M = A'*A+ 2*gamma*eye(n)
for j = 1:d    
    for i = 1:n
        if A[j,i]!=0
            omega[j] += 1               # formula for \omega_j in lecture 7
        end
    end
end
omega_max = maximum(omega)
for i = 1:3
    tau = i
    # Here, I take the min between omega and tau because in this particular example
    # the density of the matrix is .9, and tau is 1 or 2. So it's convinient to use tau
    par = min(omega_max, tau)           
    v_sparse = zeros(n,1)
    for i = 1:n
        v_sparse[i] = par*diag(M)[i]    # formula (34)
    end
    xx = norm(v_sparse)
    println("For tau = $tau the norm of v is $xx")
end
```

In this comparison, the formulation for sparse matrix and for dense matrix is the same, the reason is that this matrix is not sparse, so the $\omega_{max}$ is very big, even more than $\tau$, so at the end, when making $\min(\omega_{max}, \tau)$, it selects $\tau$. The smallest norm of the $v_i$ is achieved by the formulation (38) of lecture 7, the one corresponding to the $\tau$-nice. Also, it is interesting to say that for $\tau = 1$ the values of v in $\tau$-nice and the other formulations is equivalent. This is because in the formula (38), there is a factor $\tau-1$ that becomes 0 when $\tau = 1$, so each $v_i$ is just $M_{ii}$.


## 2) Problem 2: ESO_test

For the second question, both approaches were made. The first was to use montecarlo to find the LHS of the ESO inequality and the second approach was to find the expectation. First two functions to calculate $f(x)$ and $\nabla f(x)$ were created:

```julia
function f_calc(x, A, c_1, c_3, gamma, alpha)
    # function to compute the value of f(x)
    fx = 0
    for i = 1:d                   # For each of the summands of \phi                
        s = A[i,:]'*x             # check s = b_j^x
        if s[1]<(-1/alpha[i])     # when s < -1 (but for my formulation it is equivalent to -1/alpha)
            fx = fx + 1/2*(s + c_1[i])^2 
        elseif s[1]>(1/alpha[i])  # when s > 1 (but for my formulation it is equivalent to 1/alpha)
            fx = fx + 1/2*(s + c_3[i])^2
        else                      # when s \in (-alpha,alpha)
            fx = fx 
        end
    end
    return fx + gamma*x'*x        # add the regularization factor 
end

function df_calc(x, A, c_1, c_3, gamma, alpha, j)
    # function to compute the value of \nabda f(x) in the enty j
    dfx = zeros(1,n)
    for i = 1:d                   # For each of the summands of \phi                     
        s = A[i,:]'*x             # check s = b_j^x
        if s[1]<(-1/alpha[i])     # when s < -1 (but for my formulation it is equivalent to -1/alpha)
            dfx[j] = (dfx[j] + (s + c_1[i])*A[i,j])[1]
        elseif s[1]>(1/alpha[i])  # when s > 1 (but for my formulation it is equivalent to 1/alpha)
            dfx[j] = (dfx[j] + (s + c_3[i])*A[i,j])[1]
        else                      # when s \in (-alpha,alpha)
            dfx[j] = dfx[j]
        end
    end
    return dfx[j] + 2*gamma*x'[j] # add the regularization factor 
end
```

## Montecarlo simulation
For this part of the question, Montecarlo simulations were made to compute the LHS of the ESO formula. To test that the ESO inequality holds, one random $x$ was selected and then different samplings of size $\tau$ were generated with reservoir sampling. These samplings indicates which entries of $x$ are changed in each iteration. At the end, an average of each simulation $f(x+h_1+ ... + h_{\tau})$ was made.

```julia
function ESO_Test_MonteCarlo(A,alpha,gamma,tau)
    # Random x -----------------------------------------
    x = randn(n,1)*10                    # For this test, random entries for x will be taken
                                         # I used normally distributed random x because the 
                                         # domain is (-inf, inf) and, when multiplied by 10, 
                                         # with 95% of prob the initial values in each entry
                                         # will be between (-19.6, 19.6)
    p = zeros(n,1).+tau/n                # Probability is tau/n for each entry
    fx = f_calc(x, A, c_1, c_3, gamma, alpha)   
    dfx = zeros(n,1)  
    h = zeros(n,1)
    for i = 1:n
        dfx[i] = df_calc(x, A, c_1, c_3, gamma, alpha, i)
    end
    
    # Calculate h --------------------------------------
    
    for i = 1:n
        h[i] = -dfx[i]/v[i]              # According to the formula (3) in lecture 6
    end

    # Calculate RHS ------------------------------------
    f1 = zeros(n, 1)                    
    f2 = zeros(n, 1)
    
    for i = 1:n
        f1[i] = p[i]*dfx[i]*h[i]         # Eso inequality add two summands to f(x)  
        f2[i] = p[i]*v[i]*h[i]*h[i]      
    end

    RHS = (fx +sum(f1)+.5*sum(f2))[1]    # Formula (11) in lecture 6
    # Calculate LHS ------------------------------------ (with montecarlo)
    x_samples_vec = zeros(50000,1)       # Create a vector to save the simulations
    for i = 1:50000
        x_samples = copy(x)              # Careful when copying an array!
        selected = samplingK(tau, n)     # Use reservoir sampling (to be sure that are selected uniformly)
        for j = 1:tau                    # Given tau, update the Tau entries of x
            x_samples[selected[j]] = x_samples[selected[j]] + h[selected[j]]
        end
        x_samples_vec[i] = f_calc(x_samples, A, c_1, c_3, gamma, alpha)[1]
    end
    LHS = mean(x_samples_vec[1:50000])
    println("The LHS is: $LHS and the is RHS: $RHS")
    return RHS-LHS # Outputs the diference of the RHS-LHS (Should be positive)
end
```

```julia
# First try just to see how it works:
srand(123456)

global n = 1000              # number of columns of A
global d = 10                # number of rows of A
global B = sprandn(d,n,0.9)  # Create random normal matrix with distinct level of sparsity
global alpha = zeros(d,1)    # useful to calculate A, c_1 and c_3
global gamma = 1/d           # penalization term
global tau = 2               # parameter of the T-nice


# Because it makes 50,000 simulations for each try, it takes less than 1 minutes to run
println("For tau = 1")
tau = 1
v = ComputeESO(B, alpha, gamma, tau)
ESO_Test_vec = ESO_Test_MonteCarlo(A,alpha,gamma,tau) 

println("For tau = 2")
tau = 2
v = ComputeESO(B, alpha, gamma, tau)
ESO_Test_vec = ESO_Test_MonteCarlo(A,alpha,gamma,tau) 
```

```julia
# Now another 5 tries to test with different initial points 
# Because it makes 50,000 simulations for each try, it takes 2 minutes to run
n_tries = 5
ESO_Test_vec = zeros(n_tries,1)
sum_correct = 0
for j = 1:n_tries
    ESO_Test_vec[j] = ESO_Test_MonteCarlo(A,alpha,gamma,2)
    if ESO_Test_vec[j]>0
        sum_correct+=1
    end
end
perc_correct = sum_correct/n_tries*100
print("The percentage of random x where the ESO inequality holds is: $perc_correct")
```

As can be seen, for all the random vectors x, the MonteCarlo approximation to the LHS is less than the RHS of the ESO inequality.

### Now as the second experiment, instead of using montecarlo, the exact expectation will be calculated.

As we assumed that each sample is unifomely distributed, then the probability of each of the combinations ${n \choose \tau}$ is equal. Because of the formula of the combinations, the number can be very big. But For $\tau = 1$ and $\tau = 2$ it is feasible to compute the expected value.





```julia
using Combinatorics
# This is the same code as ESO_Test_MonteCarlo, just changes when calculating the LHS of the ESO inequality
function ESO_Test(A,alpha,gamma,tau)
    # Random x -----------------------------------------
    x = randn(n,1)*10                    # For this test, random entries for x will be taken
                                         # I used normally distributed random x because the 
                                         # domain is (-inf, inf) and, when multiplied by 100, 
                                         # with 95% of prob the initial values in each entry
                                         # will be between (-19.6, 19.6)
    p = zeros(n,1).+tau/n                # Probability is tau/n for each entry
    fx = f_calc(x, A, c_1, c_3, gamma, alpha)   
    dfx = zeros(n,1)
    h = zeros(n,1)
    for i = 1:n
        dfx[i] = df_calc(x, A, c_1, c_3, gamma, alpha, i)
    end
    
    # Calculate h --------------------------------------
    
    for i = 1:n
        h[i] = -dfx[i]/v[i]
    end

    # Calculate RHS ------------------------------------
    f1 = zeros(n, 1)                    
    f2 = zeros(n, 1)
    
    for i = 1:n
        f1[i] = p[i]*dfx[i]*h[i]
        f2[i] = p[i]*v[i]*h[i]*h[i]
    end

    RHS = (fx +sum(f1)+.5*sum(f2))[1]

    # Calculate LHS ------------------------------------ 
    x_comb = collect(combinations(1:n,tau))   # Calculate de (n tau) combinations
    sum_tot = 0
    n_temp = size(x_comb)[1]
    for i = 1:n_temp
        x_samples = copy(x)                   # Careful with copy arrays!
        if i % 100000 == 0
            println("$i combinations of $n_temp")
        end
        selected = x_comb[i]
        for j = 1:tau
            x_samples[selected[j]] = x_samples[selected[j]] + h[selected[j]]
        end
        sum_tot = sum_tot + f_calc(x_samples, A, c_1, c_3, gamma, alpha)[1]
    end
    LHS = sum_tot/n_temp             
    println("The LHS is: $LHS and the is RHS: $RHS")
    return RHS-LHS                            # Outputs the difference between RHS and LHS, shoul be positive
end
```

```julia
# First try just to see how it works for tau = 1:
srand(123456)

global n = 1000              # number of columns of A
global d = 10                # number of rows of A
global B = sprandn(d,n,0.9)  # Create random normal matrix with distinct level of sparsity
global alpha = zeros(d,1)    # useful to calculate A, c_1 and c_3
global gamma = 1/d           # penalization term
global tau = 2               # parameter of the T-nice
v = ComputeESO(B, alpha, gamma, tau)
ESO_Test_vec = ESO_Test(A,alpha,gamma,1) # For tau = 1, is very fast! just (1000 1) combinations = 1000
```

```julia
# But, for tau = 2, the combinations are much more. (It computes (1000 2) combinations = 499,500) ~2 minutes
ESO_Test_vec = ESO_Test(A,alpha,gamma,2)
```

This function just works for $\tau = 1$ and $\tau = 2$. The reason is that the number of combinations increases exponencially, so it is unfeasible to compute all. But, for these tests it works.

## 3) Problem 3: PlotSpeedup
For this problem the formula to make the funtion is:

$$\Omega(\tau) = \max_i \frac{v_i}{p_i \lambda}$$

```julia
using PyPlot
function Speedup(B, alpha, gamma, tau)
    
    v_1 = ComputeESO(B,alpha,gamma,1)
    p_1 = zeros(n,1).+1/n    
    
    v_tau = ComputeESO(B,alpha,gamma,tau)
    p_tau = zeros(n,1).+tau/n
    max_num = 0
    max_denum = 0
    
    for i = 1:n
        if (v_tau[i]/p_tau[i])>max_num
            max_num = v_tau[i]/p_tau[i]
        end
        
        if (v_1[i]/p_1[i])>max_denum
            max_denum = v_1[i]/p_1[i]
        end
    end
    
    speedup = max_num/max_denum
    return speedup
end
        


function PlotSpeedUp(B, alpha, gamma, number_tau, density, include_linear=1)
    tau_speedup = zeros(number_tau,1)
    for i = 1:number_tau
        tau_speedup[i] = Speedup(B, alpha, gamma, i)
    end
    linear = zeros(number_tau,1)
    for i = 1:number_tau
        linear[i] = 1/i
    end

    plt[:plot](1:number_tau, tau_speedup, "-", linewidth=1.0, label="Density = $density")
    if include_linear == 1
        plt[:plot](1:number_tau, linear, "--", linewidth=1.0, label="Linear speedup")
    end
    legend(loc="upper right")
    xlabel(L"\tau")
    ylabel(L"\Omega(\tau)/\Omega(1)")
    title("Speed up when tau changes")
    grid("on")
    return tau_speedup
end
```

```julia
# First try just to see how it works:
srand(123456)

density = 0.2
global B = sprandn(d,n,density)  # Create random normal matrix with distinct level of sparsity
global gamma = 1/d           # penalization term
number_tau = 40
PlotSpeedUp(B, alpha, gamma, number_tau, density, 1);

```

As the theory states, the sparse matrices are close to the linear speedup. (i.e. that for $\tau = 2$ the quotient is 1/2, for $\tau = 3$, the quotient is 1/3 and so on). ## 4) Problem 4: NSyncMinibatch

```julia
srand(12345)
x0 = randn(n,1)*100
global c_1 = zeros(d,1)    # for new_nomenclature (for the case s <-1)
global c_3 = zeros(d,1)    # for new_nomenclature (for the case s > 1)
                           # for new_nomenclature (the case s \in (-1,1) just put 0)
global A = zeros(d,n)      # for new_nomenclature

function NSyncMinibatch(B, alpha, gamma, tau, x0, k, keep_each, iter_vec)
    v = ComputeESO(B, alpha, gamma, tau);
    dfx = zeros(n,1)
    h = zeros(n,1)
    for i = 1:k
        selected = samplingK(tau, n)  # Sample uniformly (reservoir sampling)
        for j = 1:tau
            dfx[selected[j]] = df_calc(x0, A, c_1, c_3, gamma, alpha, selected[j]) # update the entry of the derivative
            h[selected[j]] = -dfx[selected[j]]/v[selected[j]] # Calculate h using formula 3 in lecture 6
            x0[selected[j]] = x0[selected[j]] + h[selected[j]]# update the vector x
        end
        fx = f_calc(x0, A, c_1, c_3, gamma, alpha)[1]
        if i % keep_each == 0
            iter_vec[round(Int,i/keep_each)] = fx
        end
    end
    return iter_vec
end
```

```julia
function plot_NSync(tau, niter, keep_each)
    x0 = ones(n, 1)*10
    niter = niter
    iter_vec = zeros(round(Int,niter/keep_each), 1)
    iter_vec = NSyncMinibatch(B, alpha, gamma, tau, x0, niter, keep_each, iter_vec)
    plt[:plot](1:round(Int,niter/keep_each), iter_vec, "-", linewidth=1.0, label="Tau = $tau")
    legend(loc="upper right")
    xlabel("Iterations (each $keep_each)")
    ylabel("f(x)")
    title("Convergence respecto to Tau")
    grid("on")
end
```

We can see from this plot that when we use a bigger $\tau$, the convergence is faster (in terms of iterations). This is because when $\tau$ increases, more entries are modified in each iteration.
